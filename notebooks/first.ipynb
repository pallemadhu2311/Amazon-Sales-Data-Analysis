```python
# Impact on Discount Percentage on Rating
data["discount_percentage"] = pd.to_numeric(data["discount_percentage"], errors="coerce")
data["rating"] = pd.to_numeric(data["rating"], errors="coerce")
discount_vs_rating = data.groupby("discount_percentage")["rating"].mean().dropna()


sns.lineplot(x=discount_vs_rating.index,y=discount_vs_rating.values,markers='' ,color='red', alpha=0.6)
plt.title("Impact of Discount Percentage on Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Average Rating")

```

```python
#Impact of Discount Percentage on Ratings
data['discount_percentage'] = pd.to_numeric(data['discount_percentage'], errors='coerce')
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')
discount_vs_rating = data.groupby("discount_percentage")["rating_count"].mean().dropna()
sns.histplot(data["discount_percentage"], bins=20, color="red", edgecolor="black", kde=True)
plt.title("Impact of Discount Percentage on Rating Count")
plt.xlabel("Discount Percentage")
plt.ylabel("Average Rating")
```

```python
# Top 10 Rating per Category
data["rating_count"] = pd.to_numeric(data["rating_count"], errors="coerce")
sales_by_category = data.groupby("category")['rating_count'].sum().reset_index()
top_10_ratings = sales_by_category.sort_values(by='rating_count', ascending=False).head(10)

sns.barplot(x="rating_count", y="category", data=top_10_ratings, palette="viridis")
plt.title('Top 10 Categories by Rating Count')
plt.xlabel('Rating Count')
plt.ylabel('Category')
```

```python
# Step 2: Create connection
engine = create_engine("mysql+pymysql://root:Madhu%40123.@127.0.0.1:3306/amazon_data")
```

```python
# Create Products Table DataFrame
products_df = df_all[['product_id', 'product_name', 'category', 'actual_price',
                      'discounted_price', 'discount_percentage', 'rating']].drop_duplicates()
```

```python
# STEP 1: Remove duplicates within the DataFrame
products_df = products_df.drop_duplicates(subset='product_id')
```

```python
# STEP 2: Fetch existing product_ids from DB
existing_ids = pd.read_sql("SELECT product_id FROM products", con=engine)
```

```python
# STEP 3: Remove already inserted ones
products_df = products_df[~products_df['product_id'].isin(existing_ids['product_id'])]
```

```python
# STEP 4: Insert only if not empty
from sqlalchemy.types import String, Text, Float
if not products_df.empty:
    products_df.to_sql('products', con=engine, if_exists='append', index=False, method='multi',
        dtype={
            'product_id': String(50),
            'product_name': Text,
            'category': String(255),
            'actual_price': Float,
            'discounted_price': Float,
            'discount_percentage': Float,
            'rating': Float
        })
    print(f"‚úÖ Inserted {len(products_df)} new products.")
else:
    print("‚ö†Ô∏è No new products to insert.")
```

```python
# Users Table
users_df = df_all[['user_id', 'user_name']].drop_duplicates()
```

```python
from sqlalchemy.types import String

# Fetch existing user_ids from the database
existing_user_ids = pd.read_sql("SELECT user_id FROM users", con=engine)

# Remove duplicates by filtering out rows with user_ids already in the database
users_df = users_df[~users_df['user_id'].isin(existing_user_ids['user_id'])]

if not users_df.empty:
    users_df.to_sql('users', con=engine, if_exists='append', index=False, method='multi',
        dtype={
            'user_id': String(50),
            'user_name': String(100)
        })
    print(f"‚úÖ Inserted {len(users_df)} new users.")
else:
    print("‚ö†Ô∏è No new users to insert.")
```

```python
# Create Reviews Table DataFrame
reviews_df = df_all[['review_id', 'product_id', 'user_id', 'review_title', 'review_content', 'rating_count']].drop_duplicates()
```

```python
# STEP 1: Remove duplicates within the DataFrame
reviews_df = reviews_df.drop_duplicates(subset='review_id')
```

```python
# STEP 2: Fetch existing review_ids from DB
existing_review_ids = pd.read_sql("SELECT review_id FROM reviews", con=engine)
```

```python
# STEP 3: Remove already inserted ones
reviews_df = reviews_df[~reviews_df['review_id'].isin(existing_review_ids['review_id'])]
```

```python
# STEP 4: Insert only if not empty
from sqlalchemy.types import String, Text, Integer
if not reviews_df.empty:
    reviews_df.to_sql('reviews', con=engine, if_exists='append', index=False, method='multi',
        dtype={
            'review_id': String(255),
            'product_id': String(255),
            'user_id': String(255),
            'review_title': Text,
            'review_content': Text,
            'rating_count': Integer
        })
    print(f"‚úÖ Inserted {len(reviews_df)} new reviews.")
else:
    print("‚ö†Ô∏è No new reviews to insert.")
```
